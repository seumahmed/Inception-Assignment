Problem-1: Temperature Check

```python
temp = float(input("Enter current temperature in Celsius: "))

if temp < 10:
    print("It's too cold!")
elif 10 <= temp <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

Problem-2: ATM Withdrawal System

```python
balance = float(input("Enter your account balance: "))
withdraw_amount = float(input("Enter amount to withdraw: "))

if withdraw_amount > balance:
    print("Insufficient Balance")
else:
    balance -= withdraw_amount
    print(f"Transaction Successful! Remaining balance: {balance}")
```

Problem-3: Student Grade Calculator

```python
Marks = int(input("Enter student marks (0-100): "))

if 90 <= Marks <= 100:
    print("Grade: A+")
elif 80 <= Marks <= 89:
    print("Grade: A")
elif 70 <= Marks <= 79:
    print("Grade: B")
elif 60 <= Marks <= 69:
    print("Grade: C")
else:
    print("Grade: Fail")
```

Problem-4: Bus Ticket Discount

```python
age = int(input("Enter passenger age: "))
ticket_price = 150  # Assuming a base price

if age < 5:
    print("Ticket Price: Free")
elif 5 <= age <= 18:
    print("Ticket Price:", ticket_price * 0.5, "(50% Discount)")
elif age >= 60:
    print("Ticket Price:", ticket_price * 0.7, "(30% Discount)")
else:
    print("Ticket Price:", ticket_price, "(Full Price)")
```

Problem-5: Restaurant Bill with Discount

```python
bill = float(input("Enter total bill amount: "))

if bill > 1000:
    discount = bill * 0.10
    final_bill = bill - discount
    print("10% discount applied! Final bill:", final_bill)
else:
    print("No discount. Final bill:", bill)
```

Problem-7: Traffic Light System

```python
light_color = input("Enter traffic light color: ").lower()

if light_color == "red":
    print("Stop")
elif light_color == "yellow":
    print("Ready to go")
elif light_color == "green":
    print("Go")
else:
    print("Invalid color")
    
```

Problem-8: Leap Year Checker

```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print(year, "is a Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

Problem-10: Electricity Bill Calculator

```python
units = int(input("Enter units consumed: "))

if units <= 100:
    total_bill = units * 5
elif units <= 200:
    total_bill = (100 * 5) + ((units - 100) * 7)
else:
    total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print("Total payable amount:", total_bill, "BDT")
```
